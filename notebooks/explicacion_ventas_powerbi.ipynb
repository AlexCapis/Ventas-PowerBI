Para ejecutar este proyecto, asegúrate de tener Power BI instalado y el conjunto de datos "ventas_videojuegos.xlsx" disponible en la carpeta "data" del repositorio.

## Visualizaciones

### 1. Total de Copias Vendidas

En esta visualización, se muestra el total de ventas en millones de copias. Esta métrica es fundamental para comprender el alcance de las ventas en general.

![Total de Copias Vendidas](../docs/images/cantidad_copias_millones.png)

### 2. Gráfico Circular de Ventas por Región

El gráfico circular muestra las ventas por región en porcentajes, lo que facilita la comparación de las contribuciones de cada región a las ventas globales. Es una herramienta poderosa para entender la distribución geográfica de las ventas.

![Gráfico Circular](../docs/images/grafico_circular.png)

### 3. Gráfico de Barras Apiladas por Plataforma y Región

Este gráfico horizontal de barras apiladas muestra las ventas por plataforma y región, lo que permite identificar las tendencias de ventas en cada plataforma en diferentes regiones. Es útil para comparar el desempeño de las plataformas en mercados específicos.

![Gráfico de Barras Apiladas](../docs/images/plataforma_por_region.png)


### 4. Gráfico de Barras Apiladas por Género y Región

En esta visualización, se exploran las ventas por género y región, lo que proporciona información sobre las preferencias de género en diferentes partes del mundo. Ayuda a identificar si ciertos géneros son más populares en ciertas regiones.

![Gráfico de Barras Apiladas](../docs/images/genero_por_region.png)

### 5. Tabla de Datos

La tabla de datos muestra información detallada sobre los videojuegos, incluyendo nombre, año, ventas, plataforma, género y editorial. Puede ser utilizada para buscar información específica sobre juegos.

![Tabla de Datos](../docs/images/tabla_ventas_videojuegos.png)

### 6. Gráfico de Columnas Apiladas por Año y Región

El gráfico de columnas apiladas vertical muestra las ventas por año y región, ayudando a identificar las tendencias de ventas a lo largo del tiempo. Es especialmente útil para visualizar cómo las ventas han evolucionado a lo largo de los años en diferentes regiones.

![Gráfico de Columnas Apiladas](../docs/images/ventas_por_ano_region.png)

## Conclusiones

En este proyecto, he llevado a cabo un exhaustivo análisis de las ventas de videojuegos utilizando Power BI, lo que ha resultado en una serie de hallazgos significativos y valiosos que arrojan luz sobre la industria de los videojuegos. A continuación, se presentan algunas conclusiones destacadas:

- **Distribución de Ventas Globales:** A través de las visualizaciones, hemos observado que las ventas de videojuegos están distribuidas de manera heterogénea en diferentes regiones del mundo. Esto sugiere que las preferencias y la demanda de los jugadores varían significativamente según la ubicación geográfica.

- **Evolución Temporal:** El gráfico de columnas apiladas por año y región revela patrones de ventas a lo largo del tiempo. Hemos identificado años en los que las ventas experimentaron un aumento significativo, lo que podría estar relacionado con el lanzamiento de consolas de videojuegos populares o títulos de juegos influyentes.

- **Impacto de las Plataformas:** El gráfico de barras apiladas por plataforma y región nos permite evaluar el rendimiento de las plataformas de videojuegos en diferentes mercados. Esto es esencial para comprender cuáles son las plataformas más populares en cada región y cómo se distribuyen las ventas en consecuencia.

- **Preferencias de Género:** El análisis de ventas por género y región proporciona información valiosa sobre las preferencias de los jugadores en términos de género de juegos. Estos datos pueden ser útiles para la toma de decisiones en la industria de los videojuegos, como la selección de títulos para diferentes regiones.

- **Facilitación de Exploración de Datos:** La incorporación de segmentaciones de datos permite a los usuarios explorar y profundizar en los datos de manera eficiente. Esto brinda la flexibilidad de examinar datos específicos según sus necesidades, lo que mejora la experiencia de usuario y la utilidad de la visualización.

- **Perspectivas de Negocio:** Este análisis puede ser de gran valor para las empresas de la industria de los videojuegos. Proporciona información estratégica sobre cómo dirigir sus esfuerzos de marketing y desarrollo de juegos para maximizar las ventas en diferentes regiones.

En resumen, este proyecto ha sido un esfuerzo significativo y de gran aprendizaje para comprender mejor el mercado de los videojuegos a través de visualizaciones y análisis de datos en Power BI. Las conclusiones obtenidas tienen el potencial de informar decisiones comerciales y estratégicas importantes en la industria de los videojuegos. Este análisis representa un valioso recurso para aquellos interesados en el mercado de videojuegos y su evolución a lo largo del tiempo.
